```python
import torch
import torch.nn as nn
from torchvision.models import resnet50

class YourModelClass(nn.Module):
    def __init__(self):
        super(YourModelClass, self).__init__()
        self.resnet = resnet50(pretrained=False, num_classes=1000)

    def forward(self, x):
        return self.resnet(x)

checkpoint_path = r'C:\d_drive\Binod_Sir_Project\resnet50_l2_eps5.ckpt'

checkpoint = torch.load(checkpoint_path, map_location=torch.device('cpu'))

model_state_dict = checkpoint['model']

model = YourModelClass()

# Remove the 'module.' prefix from the state_dict keys if present
model_state_dict = {k.replace('module.', ''): v for k, v in model_state_dict.items()}

model.load_state_dict(model_state_dict)

model.eval()
```
